```python
#1.1 creating new column
df['Age']=dateTime.year-df['year']
# 1.2 dropping year
df.drop('year',axis=1,inplace=True)
```

```python
#outlier Removal
import seaborn as sns
sns.boxplot(df['selling_price'])
```

```python
sorted(df['selling_price'],reverse=True)
```

```python
#traversing dataset for rows having outlier
df[(df['selling_price']>=8150000)&(df['selling_price']<=8900000)]
```

```python
#updating new database without outlier rows
df=df[~(df['selling_price']>=8150000)&(df['selling_price']<=8900000)]
```

```python
#Encoding categorical variable
#1
df['fuel'].unique()
```

```python
df['fuel']=df['fuel'].map({'Petrol':1,'Diesel':2,'CNG':3,'LPG':4,'Electric':5})
```

```python
#2
df['seller_type'].unique()
```

```python
df['seller_type']=df['seller_type'].map({'Individual':1,'Dealer':2,'Trustmark Dealer':3})
```

```python
#3 
df['transmission'].unique()
```

```python
df['transmission']=df['transmission'].map({'Manual':1,'Automatic':2})
```

```python
#3
df['owner'].unique()
```

```python
df['owner']=df['owner'].map({'First Owner':1,'Second Owner':2,'Fourth & Above Owner':3,'Third Owner':4,'Test Drive Car':5})
```

```python
df.head()
```

```python
 #store featur matrix X and response Y
x=df.drop(['name','selling_price'],axis=1)
y=df['selling_price']
```

```python
#Splitting dataset into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
#importing models
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
#Model training
lr=LinearRegression()
lr.fit(x_train,y_train)

rf=RandomForestRegressor()
rf.fit(x_train,y_train)

gb=GradientBoostingRegressor()
gb.fit(x_train,y_train)
```

```python
#Predicting on dataset
y_pred1=lr.predict(x_test)
y_pred2=rf.predict(x_test)
y_pred3=gb.predict(x_test)
```

```python
#Evaluating the algorithm
from sklearn.metrics import r2_score
s1=r2_score(y_test,y_pred1)
s2=r2_score(y_test,y_pred2)
s3=r2_score(y_test,y_pred3)
print(s1,s2,s3)
```
